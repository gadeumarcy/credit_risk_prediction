#### Parmi les 32 579 clients que comporte notre jeu de données, 25 472 clients ont pu rembourser leurs crédits contre 7 107 clients qui n'ont pas pu honorer leurs engagements.

#### On observe donc un déséquilibre important entre les classes : la classe "Remboursé" représente environ 78% des données, tandis que la classe "Défaut de paiement" représente seulement 22%.

#### Ce déséquilibre peut poser problème lors de l'entraînement  car celui-ci risque d'être "biaisé" en faveur de la classe majoritaire (Remboursé).

#### Pour pallier ce problème et améliorer les performances de notre modèle, notamment sur la classe minoritaire, nous avons choisi d'utiliser la méthode SMOTE (Synthetic Minority Over-sampling Technique).

#### SMOTE permet de générer artificiellement de nouvelles observations de la classe minoritaire en créant des exemples synthétiques proches des individus existants. Cela permet d'équilibrer les classes et d'entraîner un modèle plus robuste et plus juste.

```python
plt.pie(df['loan_intent']. value_counts(),labels =df['loan_intent']. value_counts().index, autopct='%1.1f%%' )
plt.show()
```

### Préparations des données pour le modèle

```python
X = df.drop("loan_status", axis =1) 
Y = df["loan_status"]
#Encoder les variables catégorielles
X = pd.get_dummies(X, drop_first=True)
colonnes = X.columns
```

### Séparation des données

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42, stratify=Y)
```

### Application du SMOTE

```python
print("Avant SMOTE :")
print(Y_train.value_counts())

# Application de SMOTE
smote = SMOTE(random_state=42)
X_train, Y_train = smote.fit_resample(X_train, Y_train)

print("Après SMOTE :")
print(Y_train.value_counts())
```

### Séparation des données  

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42, stratify=Y)
```

### Standardisation des données

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Entrainement du modèle

```python
model = XGBClassifier(random_state=42)
model.fit(X_train, Y_train)
```

### Prédiction

```python
Y_pred = model.predict(X_test)
```

### Rapport de classification

```python
print(classification_report(Y_test, Y_pred, target_names=["Remboursé", "Défaut de paiement"]))
```

### Après avoir appliqué SMOTE pour équilibrer les classes et entraîné un modèle XGBoost, les performances obtenues sont excellentes. Le modèle détecte très bien les clients remboursant leur crédit (rappel de 99%) tout en réussissant à identifier efficacement les clients à risque de défaut de paiement (rappel de 73% et précision de 95%). Ce résultat est très satisfaisant compte tenu du déséquilibre initial des classes.

### Matrice de confusion

```python
cm = confusion_matrix(Y_test, Y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Remboursé", "Défaut de paiement"])
disp.plot(cmap="Blues")
plt.title("📌 Matrice de Confusion - XGBoost")
plt.show()
```

#### La matrice de confusion ci-dessus permet d’évaluer la qualité de notre modèle de classification en observant la répartition des prédictions correctes et incorrectes.
#### Le modèle a correctement prédit 7 563 clients qui ont remboursé leur crédit (vrais négatifs).
#### Il a aussi identifié 1 551 clients comme étant en défaut de paiement (vrais positifs).
#### En revanche, 79 clients ont été faussement prédits comme étant en défaut alors qu’ils ont remboursé (faux positifs).
#### Et 581 clients réellement en défaut ont été classés à tort comme fiables (faux négatifs).

#### Ce bon équilibre entre les deux classes indique que le modèle ne se contente pas de prédire uniquement la classe majoritaire, ce qui est souvent le cas en présence de classes déséquilibrées. Grâce à l'utilisation de SMOTE, notre modèle a mieux appris les caractéristiques des clients à risque.

#### En résumé, la matrice de confusion confirme que notre modèle est à la fois précis pour détecter les bons payeurs et efficace pour repérer les clients risqués, ce qui est essentiel pour une application bancaire.

### ROC AUC

```python


probs = model.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(Y_test, probs)
plt.plot(fpr, tpr, label=f'AUC = {roc_auc_score(Y_test, probs):.2f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('Taux de Faux Positifs')
plt.ylabel('Taux de Vrais Positifs')
plt.title('Courbe ROC')
plt.legend()
plt.show()
```

#### 

#### La courbe ROC (Receiver Operating Characteristic) permet d’évaluer la capacité de discrimination de notre modèle, c’est-à-dire sa capacité à distinguer correctement les clients qui vont rembourser leur crédit de ceux qui risquent de faire défaut. Plus la courbe ROC est proche du coin supérieur gauche, meilleur est le modèle. Dans notre cas, la courbe est bien éloignée de la diagonale aléatoire, ce qui montre une bonne performance.

#### L’aire sous la courbe (AUC) est de 0.94, ce qui est très élevé. Cela signifie que dans 94 % des cas, le modèle est capable de différencier un client solvable d’un client à risque. Ce score confirme que notre modèle XGBoost, entraîné avec SMOTE pour gérer le déséquilibre des classes, est performant et fiable pour une application en milieu réel, comme la gestion du risque de crédit par une banque
